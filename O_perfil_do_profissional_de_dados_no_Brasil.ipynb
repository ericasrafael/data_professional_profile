* Para análise de salário, é ideal transformar cada categoria por um valor numérico representativo. Neste caso, fez-se a substitução de cada categoria pelo seu valor médio ( transformada em contínua ), assumindo que o valor do salário de cada pessoa se dá pelo valor central da faixa salarial em que se encontra, exceto nos extremos, que são considerados os limites.


* Também fez-se a substituição da variável número de trabalhadores, que é separada em categorias de intervalos numéricos, pelo tamanho padronizado de empresa.


* Na variável Age, fez-se uma partição por faixas etárias.

```python
# salarios

salamap = {    'Menos de R$ 1.000/mês': 1000,
      'de R$ 1.001/mês a R$ 2.000/mês': 1500, 
       'de R$ 2.001/mês a R$ 3000/mês': 2500,
      'de R$ 3.001/mês a R$ 4.000/mês': 3500,
      'de R$ 4.001/mês a R$ 6.000/mês': 5000, 
      'de R$ 6.001/mês a R$ 8.000/mês': 7000,
     'de R$ 8.001/mês a R$ 12.000/mês': 10000,
    'de R$ 12.001/mês a R$ 16.000/mês': 14000,
    'de R$ 16.001/mês a R$ 20.000/mês': 18000,
    'de R$ 20.001/mês a R$ 25.000/mês': 22500, 
              'Acima de R$ 25.001/mês': 25000 }

#tamanho da empresa por número de funcionários

tammap = { 'de 1 a 5': 'Pequena',
          'de 6 a 10': 'Pequena',
         'de 11 a 50': 'Pequena',
        'de 51 a 100': 'Pequena',
       'de 101 a 500': 'Média',
      'de 501 a 1000': 'Média',
     'de 1001 a 3000': 'Grande',
      'Acima de 3000': 'Grande'}

#categorizar idade (discretização da idade)

cortes = [0,24,30,40,100]
nomes = ['[18,24]','[25,30]','[31,40]','[41,50]']

# criando um novo dataset com as informações importantes para análise

# com o método cut (corte) do pandas, corta-se a variável age em categorias

dados["idade"] = pd.cut(dados["('P1', 'age')"] , bins = cortes , labels= nomes)

# transformando categoria de salário em variável contínua

dados['salario'] = dados["('P16', 'salary_range')"].map(salamap)  

# transformando o número de funcionários em classificação em tamanho da empresa

dados['tamanho_da_empresa'] = dados["('P12', 'workers_number')"].map(tammap) 
      
# fez-se a substituição de valores categóricos por variáveis binárias, renomeando também as variáveis para melhor entendimento

dados['gestor'] = dados["('P13', 'manager')"].map({0: 'não', 1: 'sim'})
dados['se_considera_ds'] = dados["('P19', 'is_data_science_professional')"].map({0: 'não', 1: 'sim'})
        
# mudando apenas os nomes de algumas variáveis

dados['sexo'] = dados["('P2', 'gender')"]
dados['experiencia_ds'] = dados["('P17', 'time_experience_data_science')"]
dados['tipo_de_trabalho'] = dados["('P10', 'job_situation')"]
dados['escolaridade'] = dados["('P8', 'degreee_level')"]
dados['area_de_formacao'] = dados["('D3', 'anonymized_degree_area')"]
dados['setor_de_mercado'] = dados["('D4', 'anonymized_market_sector')"]
dados['plataforma_favorita'] = dados["('P35', 'data_science_plataforms_preference')"] 
        
dados.head()
```
